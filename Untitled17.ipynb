## High-Value Customer Churn Analysis

### Subtask:
Define what constitutes a 'high-value' customer, identify them, analyze if they are churning, investigate the specific reasons or characteristics contributing to their churn, and propose potential strategies to retain them.


**Reasoning**:
To identify and analyze high-value customers who have churned, I will first calculate the 75th percentile of 'Total Revenue' from the main DataFrame to define 'high-value'. Then, I will filter the DataFrame to create `high_value_customers` and `churned_high_value_customers`. Finally, I will display the counts for both groups, along with descriptive statistics for numerical columns and value counts for key categorical columns of the `churned_high_value_customers` DataFrame to understand their characteristics and churn reasons.



```python
revenue_75th_percentile = df['Total Revenue'].quantile(0.75)
print(f"75th percentile of Total Revenue: {revenue_75th_percentile:.2f}")

# Identify high-value customers
high_value_customers = df[df['Total Revenue'] >= revenue_75th_percentile]
print(f"\nTotal number of high-value customers: {len(high_value_customers)}")

# Identify churned high-value customers
churned_high_value_customers = high_value_customers[high_value_customers['Customer Status'] == 'Churned']
print(f"\nTotal number of high-value customers who have churned: {len(churned_high_value_customers)}")

# Display descriptive statistics for numerical columns of churned high-value customers
numerical_cols = ['Age', 'Number of Dependents', 'Tenure in Months', 'Monthly Charge', 'Total Charges', 'Total Revenue']
print("\nDescriptive statistics for churned high-value customers:")
print(churned_high_value_customers[numerical_cols].describe())

# Display value counts for categorical columns of churned high-value customers
categorical_cols = ['Gender', 'Married', 'Internet Service', 'Contract', 'Payment Method', 'Churn Category', 'Churn Reason']
print("\nValue counts for churned high-value customers:")
for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(churned_high_value_customers[col].value_counts())
```

## Visualize Key Insights

### Subtask:
Generate various visualizations (e.g., bar charts for categorical features, histograms for numerical features, pie charts for proportions) to illustrate findings related to customer churn, characteristics of different customer groups, and the identified churn drivers. Ensure all plots are clearly labeled and include legends.


**Reasoning**:
To begin visualizing key insights, I will import the necessary plotting libraries, `matplotlib.pyplot` and `seaborn`, as instructed, to prepare for creating various charts.



```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries matplotlib.pyplot and seaborn imported successfully.")
```

**Reasoning**:
To visualize the overall customer status, I will create a count plot for the 'Customer Status' column from the main `df` DataFrame, adding a title and axis labels to ensure clarity.



```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Customer Status', palette='viridis')
plt.title('Overall Customer Status Distribution')
plt.xlabel('Customer Status')
plt.ylabel('Number of Customers')
plt.show()
```

**Reasoning**:
To visualize the top churn categories, I will create a count plot for the 'Churn Category' column from the `churned_customers` DataFrame, ordering the bars by frequency and adding appropriate labels and title.



```python
plt.figure(figsize=(10, 6))
sns.countplot(data=churned_customers, y='Churn Category', order=churned_customers['Churn Category'].value_counts().index, palette='magma')
plt.title('Top Churn Categories')
plt.xlabel('Number of Customers')
plt.ylabel('Churn Category')
plt.tight_layout()
plt.show()
```

**Reasoning**:
To visualize the top churn reasons, I will create a count plot for the 'Churn Reason' column from the `churned_customers` DataFrame, ordering the bars by frequency and adding appropriate labels and title.



```python
plt.figure(figsize=(12, 8))
sns.countplot(data=churned_customers, y='Churn Reason', order=churned_customers['Churn Reason'].value_counts().index, palette='viridis')
plt.title('Top Churn Reasons')
plt.xlabel('Number of Customers')
plt.ylabel('Churn Reason')
plt.tight_layout()
plt.show()
```

**Reasoning**:
To compare 'Tenure in Months' for stayed versus churned customers, I will create a histogram with KDE for both groups, using `stayed_customers` and `churned_customers` DataFrames, and add a legend for differentiation.



```python
plt.figure(figsize=(10, 6))
sns.histplot(data=stayed_customers, x='Tenure in Months', color='green', kde=True, label='Stayed', alpha=0.6)
sns.histplot(data=churned_customers, x='Tenure in Months', color='red', kde=True, label='Churned', alpha=0.6)
plt.title('Distribution of Tenure in Months for Stayed vs. Churned Customers')
plt.xlabel('Tenure in Months')
plt.ylabel('Number of Customers')
plt.legend()
plt.tight_layout()
plt.show()
```

**Reasoning**:
To compare 'Contract' types for stayed versus churned customers, I will create two subplots, each containing a count plot for the 'Contract' column from the respective `stayed_customers` and `churned_customers` DataFrames, ensuring proper labels and titles for comparison.



```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)

sns.countplot(data=stayed_customers, x='Contract', order=stayed_customers['Contract'].value_counts().index, palette='Blues', ax=axes[0])
axes[0].set_title('Contract Types for Stayed Customers')
axes[0].set_xlabel('Contract Type')
axes[0].set_ylabel('Number of Customers')

sns.countplot(data=churned_customers, x='Contract', order=churned_customers['Contract'].value_counts().index, palette='Oranges', ax=axes[1])
axes[1].set_title('Contract Types for Churned Customers')
axes[1].set_xlabel('Contract Type')
axes[1].set_ylabel('Number of Customers')

plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` regarding the usage of `palette` without explicitly assigning `hue` in `sns.countplot`. To address this, I will update the `sns.countplot` calls to explicitly set `hue` to the 'Contract' column (which is also the `x` variable) and `legend` to `False` to prevent the warning and ensure proper color mapping.



```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)

sns.countplot(data=stayed_customers, x='Contract', hue='Contract', order=stayed_customers['Contract'].value_counts().index, palette='Blues', ax=axes[0], legend=False)
axes[0].set_title('Contract Types for Stayed Customers')
axes[0].set_xlabel('Contract Type')
axes[0].set_ylabel('Number of Customers')

sns.countplot(data=churned_customers, x='Contract', hue='Contract', order=churned_customers['Contract'].value_counts().index, palette='Oranges', ax=axes[1], legend=False)
axes[1].set_title('Contract Types for Churned Customers')
axes[1].set_xlabel('Contract Type')
axes[1].set_ylabel('Number of Customers')

plt.tight_layout()
plt.show()
```

**Reasoning**:
To visualize the churn reasons specific to high-value customers, I will create a count plot for the 'Churn Reason' column from the `churned_high_value_customers` DataFrame, ordering the bars by frequency and adding appropriate labels and title.



```python
plt.figure(figsize=(12, 8))
sns.countplot(data=churned_high_value_customers, y='Churn Reason', order=churned_high_value_customers['Churn Reason'].value_counts().index, palette='plasma', hue='Churn Reason', legend=False)
plt.title('Churn Reasons for High-Value Customers')
plt.xlabel('Number of Customers')
plt.ylabel('Churn Reason')
plt.tight_layout()
plt.show()
```
